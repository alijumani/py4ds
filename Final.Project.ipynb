```python
def Winner(row):
    if row['Home Team Goals']>row['Away Team Goals']:
        return(row['Home Team Name'])
    elif row['Home Team Goals']<row['Away Team Goals']:
        return(row['Away Team Name'])
    else:
        return(row['Win conditions'])


WorldCupMatches['Winner'] = WorldCupMatches.apply(Winner, axis = 1)
# I have created a column which specifies who won that match as it will help us in some insights
```

```python
WorldCupMatches.head()
```

```python
WorldCupMatches["Winner"] = WorldCupMatches["Winner"].str.split(" ", 1, expand=True)
```

```python
WorldCupMatches.groupby('Winner')['Year'].count().sort_values(ascending=False)
# As expected, we can see that Brazil has won the most matches as they have won the most World Cups. Interestingly, Italy has won farr less, even less than Argentina who has
# won less WC's than Italy. It might indicate Italy showing up in crucial matches as well as highlights its luck
```

```python
X = WorldCupMatches[['Home Team Goals','Away Team Goals']]
```

```python
def Total_Goals_Home(x):
    return x.sum()
X.apply(Total_Goals_Home, axis = 0)
# Here we can see that throughout all the matches ever played in all the worldcups the home teams have scored nearly double the goals than the away team thus
# we can conclude that the adavantage of playing at home is immense
```

```python
def Goals(row):
    goals = row['Home Team Goals'] + row['Away Team Goals']
    return(goals)
```

```python
WorldCupMatches['Total Goals'] = WorldCupMatches.apply(Goals, axis = 1)
```

```python
# Modeling Total Goals in a match using linear regression
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
model
```

```python
X = WorldCupMatches[['Year', 'Stage', 'Home Team Name', 'Away Team Name', 'Attendance', 'Winner']]
```

```python
y = WorldCupMatches['Total Goals']
```

```python
X = pd.get_dummies(data=X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
model.fit(xTrain, yTrain)
```

```python
model.score(xTest, yTest)
# This means that the linear model is performing really bad on this dataset and fails miserably
```
